# Visualizations
In this page we are going to explore some of the available visualizations for Quantum circuits with Qiskit.
We will be using the same basic circuits for each of them.

```python
from qiskit import QuantumCircuit
from qiskit_aer import AerSimulator

qc1 = QuantumCircuit(2)
qc1.h(0)
qc1.cx(0, 1)
qc1.draw('mpl')
```

The statevector for this circuit is:

```python
from qiskit.quantum_info import Statevector

sv = Statevector(qc1)
sv.draw('latex')
```

This one is very similar to the previous one. We just add a phase so that we have slightly more interesting visualizations.

```python
qc2 = QuantumCircuit(2)
qc2.h(0)
qc2.s(0)
qc2.cx(0, 1)
qc2.draw('mpl')
```

And the statevector for this circuit is:

```python
from qiskit.quantum_info import Statevector

sv = Statevector(qc2)
sv.draw('latex')
```

## City
In order to display the histogram for our circuit in a 3D histogram, we only need to import the plot_state_city and provide the QuantumCircuit we want to visualize.  
Now we generate the visualization for the circuit:

```python
from qiskit.visualization import plot_state_city

plot_state_city(qc1)
```

```python
plot_state_city(qc2)
```

## Hinton
Another visualization which is available is the Hinton one which uses a couple of grayed out squares instead of a 3D histogram. In this case, the values are represented using different shades of grey.
With this code:

```python
from qiskit.visualization import plot_state_hinton

plot_state_hinton(qc1)
```

```python
plot_state_hinton(qc2)
```

## Pauli
The required code for the Pauli visualization is:

```python
from qiskit.visualization import plot_state_paulivec

plot_state_paulivec(qc1)
```

```python
plot_state_paulivec(qc2)
```

## QSphere
The QSphere visualization shows the output from one circuit in the bloch sphere. 
Now the required code is this:

```python
from qiskit.visualization import plot_state_qsphere

plot_state_qsphere(qc1)
```

```python
plot_state_qsphere(qc2)
```

# DAG (Directed Acyclic Graph)

DAG does not provide a way to display the expected output of a circuit but a way to display the circuit itself showing the dependencies in the circuit. DAG diagrams are used by the transpiler.
The diagram for our 'standard' circuit is:  

```python
from qiskit.visualization import dag_drawer
from qiskit.converters import circuit_to_dag

dag = circuit_to_dag(qc1)

dag_drawer(dag)
```

# Bloch Sphere

The bloch sphere plots vectors either in spherical or cartesian (default) coordinates. But does not accept a quantum circuit as a parameter, just the coordinates.

## Cartesian
This option requires the (x,y,z) coordinates.

```python
from qiskit.visualization import plot_bloch_vector
import numpy as np 
 
plot_bloch_vector([np.sqrt(2)/2,np.sqrt(2)/2,0], title="Cartesian") # cartesian coordinates by default
```

## Spherical
In this case we provide (radius, $\theta$,$\phi$) as the parameters. 
Remember that:  
$\ket{\psi} = cos(\theta/2)\ket{0}+e^{j\phi}sin(\theta/2)\ket{1}$

```python
plot_bloch_vector([1,np.pi/2,np.pi/4], title="Spherical", coord_type='spherical') 
```
